## Drop those columns

```python
# Also drop id column
df = df.drop(columns = [
    'customer_occupation_code',
    'customer_education',
    'customer_children',
    'customer_relationship', 
    'client_id'])
df.head()
```

## Handle null columns with date datatype

```python
# Fill null value with current date
cur_date = '2022-05'
df.fillna(value = cur_date, inplace = True)
```

```python
# function find months difference with date parameter is string
def diff_month(date):
    # day is 1
    cur_date = datetime(2022, 5, 1)
    date = datetime(int(date[0:4]), int(date[5:7]), 1)
    return (cur_date.year - date.year) * 12 + cur_date.month - date.month
```

```python
months_diff_all = []
months_diff_bank = []

for row in df['customer_since_all']:
    months_diff_all.append(diff_month(row))
    
for row in df['customer_since_bank']:
    months_diff_bank.append(diff_month(row))

df['customer_since_all'] = months_diff_all
df['customer_since_bank'] = months_diff_bank
df.head()
```

```python
# Replace null values (currently are 0) with mean of the attribute
mean = df[df['customer_since_all'] != 0][['customer_since_all']].mean()[0]
df = df.replace({'customer_since_all': 0}, mean)

mean = df[df['customer_since_bank'] != 0][['customer_since_bank']].mean()[0]
df = df.replace({'customer_since_bank': 0}, mean)
```

```python
# Also customer_birth_date column
customer_birth_date = []
for row in df['customer_birth_date']:
    customer_birth_date.append(diff_month(row))

df['customer_birth_date'] = customer_birth_date
df.head()
```

## Change datatype to float

```python
df = df.astype(float)
df.info()
```
